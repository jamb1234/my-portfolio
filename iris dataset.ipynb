```python
mydictionary = {"setosa":0, "versicolor":1, "virginica":2}
df["species"].map(mydictionary)
```

```python
df["species"] = df["species"].map(mydictionary)
df.species.unique()
```

```python
df.head()
```

```python
x = df[["petal_width","petal_length","sepal_width", "sepal_length"]]
y = df["species"]
```

```python
x
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 20)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
knm = KNeighborsClassifier(n_neighbors=3)
```

```python
knm
```

```python
knm.fit(x_train,y_train)
```

```python
knm.score(x_test,y_test)
```

```python
knm = KNeighborsClassifier(n_neighbors= 3)
knm.fit(x_train,y_train)
knm.score(x_test,y_test)
```

```python
n_neighbors = []
train_accuracy =[]
test_accuracy = []
range(10)
```

```python
for i in range(1,20):
    knm = KNeighborsClassifier(n_neighbors = i)
    knm.fit(x_train,y_train)
    b = knm.score(x_train,y_train)
    a = knm.score(x_test,y_test)
    train_accuracy.append(b)
    test_accuracy.append(a)
    n_neighbors.append(i)
    
```

```python
train_accuracy
```

```python
test_accuracy
```

```python
import numpy as np
```

```python
plt.plot(n_neighbors, train_accuracy,"-ro", c = "red")
plt.plot(n_neighbors, test_accuracy, "-ro", c = "blue")
plt.xlabel("Number of K-neighbors")
plt.ylabel("Accuracy score")
plt.xticks(np.arange(0,21))
plt.show()
```
